```python
import  numpy as np
import os
import glob
```

```python
my_file = open("data/part3/Wb_sRGB_output/mse_loss.txt", "r")
  
# reading the file
data = my_file.readlines()
my_dict = {}
for line in data:
    name = line.split(' ')[0]
    mse = line.split(' ')[-1].strip('\n')[1:-1]
    my_dict[name] = mse
```

```python
gt_files = glob.glob("data/part3/ground_truth/*.png")
files = glob.glob("data/part3/Wb_sRGB_output/*.jpg")
parent_path = "data/part3"

n = len(files)
i = 1
for gt_file in gt_files:
    cnt = 0
    filename = gt_file.split('/')[-1].split('_')[0]
    files = glob.glob("data/part3/Wb_sRGB_output/{}*".format(filename))
    gt_img_path = parent_path+"/ground_truth/"+filename+"_G_AS.png"
    print('<br> <table ><tbody> ')
    print('<tr><td colspan="6"> <img src="{}" width="300px"></td> </tr>'.format(gt_img_path))
    for file in files:
        if cnt == 0 :
            print('<tr>')
        name = file.split('/')[-1]
        mse = my_dict[name]
        print('<td><img src="{}" width="300px"><br><span>{}</span><br><span>{}</span></td>'.format(file, mse, name))
        if cnt == 5 :
            print('</tr>')
            cnt = -1
        cnt += 1
    print('</tbody></table> <br> ')
```
